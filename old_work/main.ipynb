```python
import csv
from pandas import DataFrame
import sklearn as sk
from sklearn import tree
from sklearn.model_selection  import  train_test_split

# personal imports
import data_extractor as de
from MultiColumnLabelEncoder import MultiColumnLabelEncoder
from Train_n_test import Train_n_test
```

# Prepare train and test datasets (+ labels)

```python
mycsv = open("data/TRAIN.CSV")
limit_seconds = 5
df, labels = de.extract_data_till_time_df(mycsv, limit_seconds)



# Get train and tes input without labels
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print("train set:\n", train_df)
#print("test set:\n", test_df)

# Get labels
train_indices = train_df.index.values
test_indices = test_df.index.values
# Recall: indices start from 0.
#print("train indices:", train_indices)
#print("test indices:", test_indices)
train_labels = [labels[i] for i in train_indices]
test_labels = [labels[i] for i in test_indices]
#print("train labels:", train_labels)
#print("test labels:", test_labels)
```

# Train and test 

```python
tnt = Train_n_test()
est = tree.DecisionTreeClassifier()
tnt.train(train_df, train_labels, est)

tnt.test(test_df, test_labels)
```

# Apply 

```python
to_estimate = test_df

tnt.estimate(to_estimate, saved_file="test_labels.CSV")
```
