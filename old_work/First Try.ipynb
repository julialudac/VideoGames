## Extracting training data : labels and features (race for this notebook)

```python
import csv



players_id_enum = [] 
races_enum = ["Zerg", "Protoss", "Terran"] # We know the races, writing it by hand will avoid some instructions
features = [] # for now, only the race => a x is one scalar/String 
labels = []


"""def find_key(mydict, value):
    for k,v in mydict.items():
        if v == value:
            return k
    return -1
"""


with open("./data/TRAIN.CSV") as csvfile:
    spamreader = csv.reader(csvfile, delimiter='\n')
    for row in spamreader:
        between_commas = row[0].split(",") # row is a 1-element table ?!
        if between_commas[0] not in players_id_enum: # if the player ID is not yet registered as a number, we add it
            players_id_enum.append(between_commas[0])
        features.append([races_enum.index(between_commas[1])])
        #print(races_enum.index(between_commas[1]))
        labels.append(players_id_enum.index(between_commas[0]))

        
for i in range(len(players_id_enum)):
    print("index " + str(i) + ": " + players_id_enum[i])
    

#for i in range(len(labels)):
#    print("Player " + labels[i] + ", aka " +  " is of race " + labels[i] " akas ")
        
        
    
```

## Constructing and feeding a Decision Tree

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(features, labels)
```

## Testing

```python
# testing on some garbage
model.predict([[0], [1], [2], [0]])

# testing on exercise test
test_features = []
with open("./data/TEST.CSV") as csvfile:
    spamreader = csv.reader(csvfile, delimiter='\n')
    for row in spamreader:
        between_commas = row[0].split(",") # row is a 1-element table ?!
        test_features.append([races_enum.index(between_commas[0])])
test_labels = model.predict(test_features)
```

### Predictions in text and saving the CSV



```python
import csv

with open("test_labeos.CSV", "w", newline="") as csvfile:
    file = csv.writer(csvfile, quoting = csv.QUOTE_NONE, delimiter=",") # we don't want of these quotes 
    file.writerow(["RowId,prediction"]) # Must put into brackets to avoid commas after each word
    for i in range(len(test_labels)):
        file.writerow([str(i+1) + "," + players_id_enum[test_labels[i]]])
```
